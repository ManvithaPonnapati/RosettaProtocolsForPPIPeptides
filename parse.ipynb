```python
import os
from glob import glob
import math
import numpy as np
import pandas as pd
import re
```

```python
big_df = pd.DataFrame()
```

```python
# original

suffix = "_0001_0001_0001"

df = pd.DataFrame()
rows = []
for filename in glob("original/score*.sc"):
    print(filename)
    with open(filename, "r") as f:
        for line in f:
            if line.startswith("SCORE:"):
                x = re.split(r' +', line.strip())
                if x[1] == 'total_score':
                    cols = [e for e in x[1:-1]]
                    continue

                name = x[-1]
                name = name[:-len(suffix)]
                
                first = [float(e) for e in x[1:7]]
                last = [float(e) for e in x[-50:-1]]
                row = [name] + first + last

                rows.append(row)
    print(len(rows))
            
cols = ['name'] + cols[:6] + cols[-49:]
df = pd.DataFrame(rows, columns=cols).sort_values(by=['name'])
df = df[["name", "total_score", "I_bsa", "I_sc" , "pep_sc"]]
# aggregation_functions = {'total_score': 'mean', 'I_bsa': 'mean', 'I_sc': 'mean', 'pep_sc' : 'mean'}
# df = df.groupby(df['name']).aggregate(aggregation_functions)
df = df.set_index('name', verify_integrity=True)

big_df = pd.concat([big_df, df], axis=1, join='outer', sort=True)
big_df
```

```python
# solubility

out = []
with open("solubility/seq_prediction_peptides.txt") as f:
    for line in f:
        if line.startswith("SEQUENCE PREDICTIONS"):
            x = line.split(",")
            # x[4] - population is equal for all peptides
            norm_sol = float(x[3]) - 0.5
            out.append([x[1][1:], norm_sol])
            
df = pd.DataFrame(out, columns = ['name', 'solubility'])
df = df.sort_values(by=['name'])
df = df.set_index('name', verify_integrity=True)
df

big_df = pd.concat([big_df, df], axis=1, join='outer', sort=True)
big_df
```

```python
# old spike protein

prefix = "pdb_scores_old_spike_flex_pep_dock/2ajf-"
suffix = "_B_0001_0001.pdbscore.sc"

df = pd.DataFrame()
for filename in glob("pdb_scores_old_spike_flex_pep_dock/*.sc"):
    name = filename[len(prefix):]
    name = name[:-len(suffix)]
    with open(filename, "r") as f:
        for line in f:
            if line.startswith("SCORE:"):
                x = re.split(r' +', line.strip())
                if x[1] == 'total_score':
                    cols = [e for e in x[1:-1]]
                    continue
                row = [float(e) for e in x[1:-1]]
                cols = ["old_spike_" + col for col in cols]
                temp_df = pd.DataFrame([[name] + row], columns=['name'] + cols)
                
                cols_to_drop = []
                for col in cols:
                    if col.startswith("old_spike_bestRMS_"):
                        cols_to_drop.append(col)
                temp_df.drop(cols_to_drop, axis=1, inplace=True)
                
                df = pd.concat([df, temp_df])

df = df.sort_values(by=['name'])
df = df[["name", "old_spike_total_score", "old_spike_I_bsa", "old_spike_I_sc" , "old_spike_pep_sc"]]
df = df.set_index('name', verify_integrity=True)

big_df = pd.concat([big_df, df], axis=1, join='outer', sort=True)
big_df
```

```python
#integrin

suffix = "_B_0001_0001_0001"

df = pd.DataFrame()
rows = []
for filename in glob("integrin/score_min_score*.sc"):
    print(filename)
    with open(filename, "r") as f:
        for line in f:
            if line.startswith("SCORE:"):
                x = re.split(r' +', line.strip())
                if x[1] == 'total_score':
                    cols = [e for e in x[1:-1]]
                    continue

                name = x[-1].split("-", 1)[1]
                name = name[:-len(suffix)]
                
                first = [float(e) for e in x[1:7]]
                last = [float(e) for e in x[-50:-1]]
                row = [name] + first + last

                rows.append(row)
    print(len(rows))
            
cols = ['name'] + cols[:6] + cols[-49:]
cols = ['integrin_' + col for col in cols]
df = pd.DataFrame(rows, columns=cols).sort_values(by=['integrin_name'])
df = df[["integrin_name", "integrin_total_score", "integrin_I_bsa", "integrin_I_sc" , "integrin_pep_sc"]]
aggregation_functions = {'integrin_total_score': 'mean', 'integrin_I_bsa': 'mean', 'integrin_I_sc': 'mean', 'integrin_pep_sc' : 'mean'}
df = df.groupby(df['integrin_name']).aggregate(aggregation_functions)

big_df = pd.concat([big_df, df], axis=1, join='outer', sort=True)
big_df
```

```python
aa = []
for name in list(big_df.index):
    m = re.search('receptorE_partnerA_(\d+)aa', name)
    aa.append(int(m.group(1)))
big_df["AA length"] = aa
```

```python
big_df.to_csv("final.csv")
```
